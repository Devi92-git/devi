####Question 31: Check if string is a valid email format: "user@example.com"

```python
user_input = "user@example.com"

parts = user_input.split('@')

if len(parts) != 2:
    print("Invalid email")
else:
    username, domain = parts

    if username == "" or domain == "":
        print("Invalid email")
    elif '.' not in domain:
        print("Invalid email")
    else:
        domain_parts = domain.split('.')
        if "" in domain_parts:
            print("Invalid email")
        else:
            print("Valid email")
```

####Question 41: Check if string contains balanced parentheses: "((()))"

```python
text = "((()))"
balance = 0
is_balanced = True

for ch in text:
    if ch == '(':
        balance += 1
    elif ch == ')':
        balance -= 1

    if balance < 0:
        is_balanced = False
        break

if balance != 0:
    is_balanced = False

if is_balanced:
    print("Balanced parentheses")
else:
    print("Not balanced")
```

####Question 42: Convert "hello world" to Morse code

```python
text = "hello world"
text = text.lower()

morse_code = {
    'a': '.-',   'b': '-...', 'c': '-.-.', 'd': '-..',
    'e': '.',    'f': '..-.', 'g': '--.',  'h': '....',
    'i': '..',   'j': '.---', 'k': '-.-',  'l': '.-..',
    'm': '--',   'n': '-.',   'o': '---',  'p': '.--.',
    'q': '--.-', 'r': '.-.',  's': '...',  't': '-',
    'u': '..-',  'v': '...-', 'w': '.--',  'x': '-..-',
    'y': '-.--', 'z': '--..',
    ' ': '/'
}

result = ""

for ch in text:
    result += morse_code[ch] + " "

print(result.strip())
```

####Question 44: Check if string is a valid URL: "https://www.google.com"

##### Question 46: Convert "hello world" to Pig Latin

```python
sentence = 'hello world'
arr = sentence.split()
result = []

vowels = "aeiou"

for word in arr:
    if word[0].lower() in vowels:
        result.append(word + "way")
    else:
        result.append(word[1:] + word[0] + "ay")

print(" ".join(result))
```

####Question 47: Check if string is a valid IPv4 address: "192.168.1.1"

```python
address = "192.168.1.1"
cleaned_address = address.split(".")

flag = True

if len(cleaned_address) != 4:
    flag = False
else:
    for word in cleaned_address:
        if not word.isdigit():
            flag = False
            break
        elif not (0 <= int(word) <= 255):
            flag = False
            break

print("Valid IPv4 address" if flag else "Invalid IPv4 address")
```

#### Question 49: Convert "hello world" to ROT13 encoding

```python
text = "hello world"
result = ""

for ch in text:
    if ch.isalpha():

        offset = ord(ch) - ord('a')

        new_char = chr((offset + 13) % 26 + ord('a'))
        result += new_char
    else:
        result += ch

print(result)
```

####Question 50: Check if string is a valid credit card number: "4532015112830366"

```python
card_number = "4532015112830366"


if len(card_number) != 16 or not card_number.isdigit():
    print("Invalid credit card number")
else:
    total = 0
    i = 0

    for ch in card_number[::-1]:
        digit = int(ch)

        if i % 2 == 1:
            digit *= 2
            if digit > 9:
                digit -= 9

        total += digit
        i += 1


    if total % 10 == 0:
        print("Valid credit card number")
    else:
        print("Invalid credit card number")
```
