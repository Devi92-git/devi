#####  Question 9: Find the greatest common divisor (GCD) of 48 and 36

```python
a = 48
b = 36

gcd = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i

print(gcd)
```
