```python
import numpy as np
```

```python
data = np.array([10, 20, np.nan, 40, 50, np.nan, 70])
```

```python
data
```

```python
print("original data:",(data))
```

```python
print("null positions:",np. isnan(data))
```

```python
print("total null values:",np.isnan(data).sum())
```

```python
clean_data = data[~np.isnan(data)]
print("after removing nulls:",clean_data)
```

```python
replace_zero = np.nan_to_num(data, nan=0)
print("Replace NAN with 0:",replace_zero)
```

```python
mean_value = np.nanmean(data)
```

```python
replace_mean = np.where(np.isnan(data),mean_value,data)
```

```python
print("replace nan with name:",replace_mean)
```

```python
median_value = np.nanmedian(data)
```

```python
replace_median = np.where(np.isnan(data),median_value,data)
```

```python
print("replace nan with median:",replace_median)
```
